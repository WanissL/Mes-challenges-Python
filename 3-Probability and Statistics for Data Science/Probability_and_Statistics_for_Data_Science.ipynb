### Objective

# Function Method :

```python
# Function method:
def analyse_weather_data(filename):
      """Analyzes weather data from a given file.
    Args: 
        Filename: The name of the weather data file.
    Returns:
        A tuple containing:
            - Average maximum temperature
            - Average minimum temperature
            - Total precipitation
            - Number of rainy days
            - Maximum temperature
            - Minimum temperature"""
      with open(filename,'r') as file:
        lines = file.readlines()
      lines=[line.strip() for line in lines]
      max_temp=[]
      min_temp=[]
      prec=[]
      rainy_days=0
      for i in lines:
              max_temp.append(float(i.split(',')[1]))
              min_temp.append(float(i.split(',')[2]))
              prec.append(float(i.split(',')[3]))
              rainy_days=sum([1 for j in prec if j>0])
      average_max=(sum(max_temp))/len(max_temp)
      average_min=(sum(min_temp))/len(min_temp)
      min_t=min(min_temp)

      max_t=max(max_temp)
      total_prec=sum(prec)
      return {"The average maximum temperature is " : average_max,
              "The average minimum temperature is ": average_min,
              "The total of precipitaion is ":total_prec,
              "The number of rainy days is ": rainy_days,
              "The highest temperature  recorded is ": max_t,
              "The lowest temperature  recorded is ":min_t}
```

```python
analyse_weather_data("weather_data.txt")
```
